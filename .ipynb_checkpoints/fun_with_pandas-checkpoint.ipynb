## Music sales challenge

Write a script that:

1. Combines the tables of best-selling physical singles and best-selling digital singles on the Wikipedia page "List_of_best-selling_singles"
2. Adds a column which marks whether each row is from the list of physical singles or digital singles
3. Outputs the artist and single name for the year you were born. If there is no entry for that year, take the closest year after you were born.
4. Outputs the artist and single name for the year you were 15 years old.

```python
# Starter code...

%pip install lxml
import pandas as pd

rawdata = pd.read_html("https://en.wikipedia.org/wiki/List_of_best-selling_singles")
df_phy = rawdata[0]
df_dig = rawdata[3]

df_phy
```

```python
df_dig
```

```python
# Add column which marks whether each row is from the list of physical singles or digital singles

# Add a column indicating the source
df_phy['source'] = 'physical'
df_dig['source'] = 'digital'

# Combine the tables
df_comb = pd.concat([df_phy, df_dig])

df_comb
```

```python
# Outputs the artist and single name for the year you were born.
my_year = df_comb.loc[14, ['Artist', 'Single']]

# Output the result
print(my_year)
```

## Space challenge

1. Make a single dataframe that combines the space missions from the 1950's to the 2020's
2. Write a script that returns the year with the most launches
3. Write a script that returns the most common month for launches
4. Write a script that ranks the months from most launches to fewest launches


```python
# Starter code...

#%pip install lxml
#import pandas as pd

rawdata_ = pd.read_html("https://en.wikipedia.org/wiki/Timeline_of_Solar_System_exploration")
```

```python
# Make a single dataframe that combines the space missions from the 1950's to the 2020's

df_space = pd.concat (rawdata_[0:8])
df_space
```

```python
#Write a script that returns the year with the most launches

# Step 1: Convert the 'launch_date' column to datetime format
df_space['Launch date'] = pd.to_datetime(df_space['Launch date'], format='%d %B %Y')

# Step 2: Extract the year and month from 'Launch date' and create new columns
df_space['year'] = df_space['Launch date'].dt.year
df_space['month'] = df_space['Launch date'].dt.month_name()

# Step 3: Find the year with the most missions
most_common_year = df_space['year'].mode()[0]

# Step 4: Find the most common month for missions
most_common_month = df_space['month'].mode()[0]

# Output the results
print("Year with the most missions:", most_common_year)
print("Most common month for missions:", most_common_month)
```

```python
#Write a script that ranks the months from most launches to fewest launches
# Rank the months based on the number of missions
month_ranking = df_space['month'].value_counts().sort_values(ascending=False)

# Output the ranking
print("Ranking of months by number of missions:")
month_ranking
```

## Stroop challenge

Every year between 2015 and 2021, the students in my Language, Cognition, and the Brain course participated in a version of the Stroop task. Using a stopwatch (ok, using their phones), they recorded how fast they could say a list of things (either reading or naming colors or color words). The column names mean "Reading with No Interference", "Naming with Interference", "Naming with No Interference", and "Reading with Interference". The times are in seconds.

### Stroop challenge 1: 
Transform these data from wide format to long format, so that the result is a dataframe with
- 1 column named "Participant_id" with a unique number for each participant (you can use the row indices)
- 1 column named "Year" with the year data
- 1 column named "Task" that shows which task they were doing
- 1 column named "RT" that shows their response time

```python
# Starter code 1...

df = pd.read_csv("https://raw.githubusercontent.com/ethanweed/Stroop/master/Stroop-raw-over-the-years.csv")
df
```

```python
# Make a new column using the dataframe indices as participant numbers
df.index.name = 'Participant_id'
df = df.reset_index()

# Melt the DataFrame from wide to long format
df_long = pd.melt(
    df, 
    id_vars=["Participant_id", "Year"],  # Columns to keep ("Participant_id" and "Year")
    var_name="Task",   # New column name for task type (based on original column names)
    value_name="RT"    # New column name for response times
)

# Reorder columns to match the desired output
df_long = df_long[["Participant_id", "Year", "Task", "RT"]]

# Print the transformed DataFrame
df_long
```

## Stroop challenge 2 (Advanced!!!):

Make a new dataframe which shows the mean response time (in seconds) for each task for each year.

```python
df_new = df_long.groupby(['Task', 'Year'])['RT'].mean().round(2).to_frame()
df_new
```
